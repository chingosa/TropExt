#### Motivation:
- Extreme temps in the tropics - a first order impact of climate change
- Recent extreme heat events seem to be breaking records left and right
- Extremes in the extratropics seem well established - tropics we have this idea that dry get hotter due the partitioning of more energy into sensible heat flux

Additionally...
- CMIP6 models are a little all over the place when it comes to where they think extreme temps are going to go under warming scenarios

#### The big question...
**Can we leverage a part of Mikes dry get hotter theory to provide a constraint on future high percentile temperatures**

---
#### The theory: 

#### $$\delta T_{L,fix\; RH}^x = \gamma^{T_O} \delta T_O$$

- Fixed Relative humidity version of the theory
- States the change in temperature of all days exceeding the xth percentile of temperature is equal to the change in ocean temperature for the percentile of MSE corresponding to the average MSE of days exceeding the xth percentile of temperature in the historical simulation * *deep breath* * times some sensitivity parameter gamma which scales $ \delta T_O$ 

In Mikes 2021 paper he found that these look like they should be pretty well correlated:

#### Emergent constraint-iness
##### What is an emergent constraint:
- defined as physically explainable empirical relationships between characteristics of the **current** climate and long term climate predictions that emerge in collections of climate model simiulations
- Two main things you need - convincing correlation between a variable that can be measured in the current climate and a response variable in the future climate
- for us that would be $ \gamma^{T_O}$ (which can mostly be calculated from current climate) and $\delta T_{L}^x$
