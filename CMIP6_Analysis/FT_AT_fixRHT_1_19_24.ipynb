```python
def ProcessStage2_GC(model, experiment):
        quantiles = np.array([0.001, 1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 82.5, 85, 87.5, 90, 92.5, 95, 97.5, 99])/100
        Vars = ['ps', 'tas', 'huss']

        if experiment == '4x': pList = ['piControl', 'abrupt-4xCO2']
        elif experiment == 'SSP245': pList = ['historical', 'ssp245']
        
        for period  in pList:
            savePath = f'TempData/{model}_{period}_1_10_25_GC.nc'
            if not os.path.exists(savePath):
                ds = func.grabDS(model, period, Vars)
                ds = func.AddLandMask(ds)
                ds = ds.expand_dims(model = [model])
                ds = ds.expand_dims(period = [period])
                ds = func.CalcSatSpecificHumidity(ds)
                ds = func.CalcMSE(ds)
                
                ds = ds.chunk({"time": -1})
                
                ## Grid Cell Specific Calculations
                ds = func.meanOverPercentile(ds, 'tas', ['tas', 'huss', 'qsat', 'MSE'], quantiles, ['time'], additionalMask = (ds.landseamask == 1), add_brev = '_land')
                
                MSE_bin_edges = func.generatePDF_bins(ds.MSE, 500)
                
                ds[f'MSE_pdf'] = func.generatePDF(ds.MSE.where(ds.landseamask == 1), MSE_bin_edges, dimOver = ['time'])
                ds['MSE_land_px']  = func.estimatePx(ds.MSE_pdf, ds.MSE_MOP_land)
                    
                    
                for Var in ['tas', 'huss', 'qsat']:
                    if period == pList[1]:  # if its the change period - ie 4xCO2 or SSP245
                        baseSim_land_px = xr.open_dataset(f'TempData/{model}_{pList[0]}_1_10_25_GC.nc').MSE_land_px
                        baseSim_land_px = baseSim_land_px.assign_coords({"period": [pList[1]]})
                        px = baseSim_land_px
                    else: 
                        px = ds.MSE_land_px
                        
                    bin_edges = func.generatePDF_bins(ds[Var], 500)
                    ds[f'{Var}_pdf'] = func.generatePDF(ds[Var].where(ds.landseamask == 0), bin_edges, dimOver = ['time', 'lon'])
                    ds[f'{Var}_ocean'] = func.ValuefromPx(ds[f'{Var}_pdf'], px) # tas_pdf is lat, and bins       ds_MSE_land_px is lat lon and quantile over land only
                    
                    
                    
                    ds[f'{Var}_land_mean'] = ds[Var].where(ds.landseamask == 1).mean('time')
                    
                ds = ds.drop_vars(['time', 'height', 'huss', 'ps', 'tas', 'qsat', 'MSE', 'MSE_pdf', 'MSE_bins' 'tas_qant', 'bins_tas', 'tas_pdf', 'bins_huss', 'huss_pdf', 'bins_qsat', 'qsat_pdf'], errors = 'ignore')
                
                ds.to_netcdf(savePath)
        # for Var in ['tas', 'huss', 'ps']: 
        #     if os.path.exists(f'TempData/{model}_{period}_{Var}_processed.nc'): 
        #         os.remove(f'TempData/{model}_{period}_{Var}_processed.nc')



def ProcessStage2_LW(model, experiment):
    quantiles = np.array([0.001, 1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 82.5, 85, 87.5, 90, 92.5, 95, 97.5, 99])/100
    Vars = ['ps', 'tas', 'huss']

    if experiment == '4x': pList = ['piControl', 'abrupt-4xCO2']
    elif experiment == 'SSP245': pList = ['historical', 'ssp245']
        
    for period  in pList:
        savePath = f'TempData/{model}_{period}_1_10_25_LW.nc'
        if not os.path.exists(savePath):
            ds = func.grabDS(model, period, Vars)
            ds = func.AddLandMask(ds)
            ds = ds.expand_dims(model = [model])
            ds = ds.expand_dims(period = [period])
            ds = func.CalcSatSpecificHumidity(ds)
            ds = func.CalcMSE(ds)
            
            ds = ds.chunk({"time": -1})
            ds = func.meanOverPercentile(ds, 'tas', ['tas', 'huss', 'qsat', 'MSE'], quantiles, ['time', 'lon'], additionalMask = (ds.landseamask == 1), add_brev = '_land')
            
            
            MSE_bin_edges = func.generatePDF_bins(ds.MSE, 500)
            
            ds[f'MSE_pdf'] = func.generatePDF(ds.MSE.where(ds.landseamask == 1), MSE_bin_edges, dimOver = ['time', 'lon'])
            ds['MSE_land_px']  = func.estimatePx(ds.MSE_pdf, ds.MSE_MOP_land)
            
                
                
            for Var in ['tas', 'huss', 'qsat']:
                if period == pList[1]:  # if its the change period - ie 4xCO2 or SSP245
                    baseSim_land_px = xr.open_dataset(f'TempData/{model}_{pList[0]}_1_10_25_LW.nc').MSE_land_px
                    baseSim_land_px = baseSim_land_px.assign_coords({"period": [pList[1]]})
                    px = baseSim_land_px
                else: 
                    px = ds.MSE_land_px
                    
                bin_edges = func.generatePDF_bins(ds[Var], 500)
                ds[f'{Var}_pdf'] = func.generatePDF(ds[Var].where(ds.landseamask == 0), bin_edges, dimOver = ['time', 'lon']) # in LW this is a pdf for each lat of what the ocean is up to
                ds[f'{Var}_ocean'] = func.ValuefromPx(ds[f'{Var}_pdf'], px, latWise = True) # it looks at the pdf of var over the ocean for a given lat and calcs a value looking at px - the px here is generalized by lat - in gc its specific to the gridcell
                ds[f'{Var}_land_mean'] = ds[Var].where(ds.landseamask == 1).mean('time')
                
            ds = ds.drop_vars([
             'time',
             'height',
             'huss',
             'ps',
             'tas',
             'qsat',
             'MSE',
             'MSE_pdf',
             'MSE_bins'
             'tas_qant',
             'bins_tas',
             'tas_pdf',
             'bins_huss',
             'huss_pdf',
             'bins_qsat',
             'qsat_pdf'], errors = 'ignore')
            
            ds.to_netcdf(savePath)
            # for Var in ['tas', 'huss', 'ps']: 
            #     if os.path.exists(f'TempData/{model}_{period}_{Var}_processed.nc'): 
            #         os.remove(f'TempData/{model}_{period}_{Var}_processed.nc')
            
```
