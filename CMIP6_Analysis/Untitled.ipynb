```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
sd_original=1
sd_shifted=1
# Generate data for the original and shifted normal distributions
s = 6

x = np.linspace(-s * sd_original, s* sd_original, 500)
y = (1 / (sd_original * np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x / sd_original) ** 2)

x_shifted = np.linspace(-s * sd_shifted + 2, s * sd_shifted + 2, 500)
y_shifted = (1 / (sd_shifted * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_shifted - 2) / sd_shifted) ** 2)

# Create the plot
plt.figure(figsize=(8, 5), facecolor='white')
plt.plot(x, y, color='black', lw=2, label=f'Original (SD={sd_original})')
plt.plot(x_shifted, y_shifted, color='black', lw=2, label=f'Shifted (SD={sd_shifted})')

# Add an arrow from the peak of the original to the shifted peak
plt.annotate('', xy=(2, max(y_shifted)), xytext=(0, max(y)),
             arrowprops=dict(facecolor='black', arrowstyle='->', lw=1.5))

# Bold formatting for title and labels
plt.title('Shifted Normal Distribution with Adjustable SD', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=14, fontweight='bold')

# Remove axis ticks
plt.xticks([])
plt.yticks([])
array1 = np.linspace(0, 0.02, 90)
array2 = np.linspace(0.02, 0, 90)

# Create a smoother transition at the peak
buffer = np.linspace(0.019, 0.02, 20)
smooth_top = np.concatenate((array1, buffer, array2))
smooth_top_smoothed = gaussian_filter(smooth_top, sigma=3)  
plt.plot(x_shifted[-200:], y_shifted[-200:]+smooth_top_smoothed, color = 'black', ls = ':')
# Show the plot
plt.show()

```
